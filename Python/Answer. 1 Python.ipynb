Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word.

**Note** - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.

**Example** input - string = ‚Äúwrite write write all the number from from from 1 to 100‚Äù.

**Example** output - 5

**Explanation** - From the given string we can note that the most frequent words are ‚Äúwrite‚Äù and ‚Äúfrom‚Äù and
the maximum value of both the values is ‚Äúwrite‚Äù and its corresponding length is 5.



```python
from collections import Counter

def find_highest_frequency_word_length(string):
    # Split the string into words
    words = string.split()

    # Count the frequency of each word using Counter
    word_frequency = Counter(words)

    # Find the word with the highest frequency
    highest_frequency_word = max(word_frequency, key=word_frequency.get)

    # Return the length of the word with the highest frequency
    return len(highest_frequency_word)


# Test cases

# Test case 1
string2 = "the quick brown fox jumps over the lazy dog"
print("Highest frequency word length: ",find_highest_frequency_word_length(string2)) 
# Test case 2
string3 = "apple apple banana banana banana cherry"
print("Highest frequency word length: ",find_highest_frequency_word_length(string3))  
```

```python
test_cases = [
  (string2, 3),
  (string3, 6)
]

for input_string, expected_output in test_cases:
  actual_output = find_highest_frequency_word_length(input_string)
  if actual_output == expected_output:
    print("Test case passed.")
  else:
    print("Test case failed.")
```

**Test Case 1**

Expected Output: 3

Explanation: In test case 1, the word "the" appears twice in the input string, which is the highest frequency. Therefore, the correct output is 3, the length of the word "the".

**Test Case 2**
