```python
Link = "https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD"
```

```python
import requests
import csv
import pandas as pd
```

```python
def get_data_as_df(link):
    with requests.get(link, stream=True) as r:
        lines = (line.decode('utf-8') for line in r.iter_lines())
        data = []
        for row in csv.reader(lines):
            data.append(row)
    df = pd.DataFrame(data)
    df.columns = df.loc[0]
    df.drop(index=0, axis=0, inplace=True)
    return df
```

```python
df = get_data_as_df(Link)
df.head()
```

```python
import plotly.graph_objects as go

# Calculate the count of electric vehicle types by city
vehicle_type_count = df.groupby("City")["Electric Vehicle Type"].count().reset_index()

# Create the bar plot using Plotly
fig = go.Figure(data=[go.Bar(x=vehicle_type_count["City"], y=vehicle_type_count["Electric Vehicle Type"])])

# Update the layout
fig.update_layout(
    title="Distribution of Electric Vehicle Types by City",
    xaxis_title="City",
    yaxis_title="Count"
)

# Show the plot
fig.show()
```
