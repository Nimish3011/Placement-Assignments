# **Task 1: Get all the overall ratings for each season and compare them using plots**

```python
import pandas as pd
import plotly.express as px

# Read the data into a DataFrame
data = pd.read_csv('westworld_episodes.csv')

# Calculate average ratings for each season
season_ratings = data.groupby('Season')['Average Rating'].mean().reset_index()

# Create a bar plot for the average ratings
fig = px.bar(season_ratings, x='Season', y='Average Rating', title='Average Ratings for Each Season')
fig.show()

```

# Task 2: Get all the episode names whose average rating is more than 8 for every season

```python
# Filter episodes with average rating > 8 for each season
high_rated_episodes = data[data['Average Rating'] > 8]

# Group episode names by season
episode_names_high_rated = high_rated_episodes.groupby('Season')['Name'].apply(list).reset_index()

print(episode_names_high_rated)
```

# **Task 3: Get all the episode names that aired before May 2019**

```python
# Convert Airdate column to datetime
data['Airdate'] = pd.to_datetime(data['Airdate'])

# Filter episodes aired before May 2019
before_may_2019_episodes = data[data['Airdate'] < pd.to_datetime('2019-05-01')]

# Group episode names by season
episode_names_before_may_2019 = before_may_2019_episodes.groupby('Season')['Name'].apply(list).reset_index()

print(episode_names_before_may_2019)
```

# **Task 4: Get the episode name from each season with the highest and lowest rating**

```python
# Get the index of episodes with the highest and lowest rating for each season
highest_rated_episodes = data.groupby('Season')['Average Rating'].idxmax()
lowest_rated_episodes = data.groupby('Season')['Average Rating'].idxmin()

# Retrieve the episode names based on the index
highest_rated_names = data.loc[highest_rated_episodes, ['Season', 'Name']].reset_index(drop=True)
lowest_rated_names = data.loc[lowest_rated_episodes, ['Season', 'Name']].reset_index(drop=True)
```

# Task 5: Get the summary for the most popular (ratings) episode in every season

```python
# Get the index of the most popular episode (highest rating) for each season
most_popular_episodes = data.groupby('Season')['Average Rating'].idxmax()

# Retrieve the summary for the most popular episodes
summary_most_popular = data.loc[most_popular_episodes, ['Season', 'Name', 'Summary']].reset_index(drop=True)
```
